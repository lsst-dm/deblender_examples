```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
# Calculate the radial distance to each pixel
max_size = 1001
x = np.arange(max_size)
y = np.arange(max_size)
X,Y = np.meshgrid(x,y)
cy = cx = max_size >> 1
X = X - cx
Y = Y - cy
r = np.sqrt(X**2+Y**2)
# Sort the pixels by radius
radial = np.argsort(r.flatten())
order = np.argsort(radial).reshape(max_size, max_size)
inf = X==0
_X = X.copy()
_X[inf] = 1
angles = np.arctan2(-Y, -_X)
angles[inf&(Y!=0)] = 0.5*np.pi*np.sign(angles[inf&(Y!=0)])
```

```python
plt.imshow(angles)
plt.show()
```
